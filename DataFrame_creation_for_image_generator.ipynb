```
import numpy as np 
import cv2
import os
from os import listdir
from os.path import isfile, join
from google.colab import files
import csv
import pandas as pd
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
```

```
# Get the file names in a list(All three sets of images stored in one folder)
mypath='/content/gdrive/My Drive/Autoencoder_Dataset/Depth/Depth_autoenco/'
onlyfiles=[]
list_dir=os.listdir(mypath)
#print(list_dir)
for f in list_dir:
  onlyfiles.append((mypath+f))
print(len(onlyfiles))
print(onlyfiles)

```

```
# Creation of CSV files containing path, target value and coordinates of all the images
with  open('dataset_info_autoencoder.csv','w') as file:
  #count=1
  writer = csv.writer(file)
  writer.writerow(["Filename","Path","Target","x1","y1","x2","y2"])
  for i in onlyfiles:
    j=i.split('/')[-1]
    target_value_string = j.split('_')[4].split('_')[0]
    #print(j)
    #print(target_value_string)
    x1=j.split('_')[5].split('_')[0]
    y1=j.split('_')[6].split('_')[0]
    x2=j.split('_')[7].split('_')[0]
    y2=j.split('_')[8].split('.')[0]
    #print(x1)
    #print(y1)
    #print(x2)
    #print(y2)
    writer.writerow([j,i,target_value_string,x1,y1,x2,y2])
    #count=count+1
files.download('dataset_info_autoencoder.csv')
```

```
# Creating Train and Test DataFrames from the CSV file created in the previous step
df=pd.read_csv('file_info1.csv')
#df1=df[['S No','Filename','Path']]
X_train, X_test, y_train, y_test = train_test_split(df, df, test_size=0.2,random_state=1)
#print(len(X_train))
#print(len(X_test))
#print(X_train)


# Converting the obtained train and test dataframes as CSV files 
X_train.to_csv('train_info_autoencoder.csv',index=False)
X_test.to_csv('test_info_autoencoder.csv',index=False)
files.download('train_info_autoencoder.csv')
files.download('test_info_autoencoder.csv')
```
